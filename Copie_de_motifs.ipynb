Exercice 3: Toujours avec une **boucle for**, dessiner le motif suivant en utilisant des **espaces ' '** et des **étoiles '*'**::

```python
'''
    *
   **
  ***
 ****
*****
'''
```

```python
'''
    *
   **
  ***
 ****
*****
'''
dimension = 5
for i in range(1, dimension + 1):
    print(' ' * (dimension - i) + '*' * i)
```

Exercice 4: Toujours avec une **boucle for**, dessiner le motif suivant en utilisant des **espaces ' '** et des **étoiles '*'**:

```python
'''
    *
   ***
  *****
 *******
*********
'''
```

```python
'''
    *
   ***
  *****
 *******
*********
'''
dimension = 5
for i in range(1, dimension + 1):
    print(' ' * (dimension - i) + '*' * (2 * i - 1))
```

Exercice 5: Toujours avec une **boucle for**, dessiner le motif suivant en utilisant des **espaces ' '** et des **étoiles '*'**:

```python
'''
*********
 *******
  *****
   ***
    *
'''
dimension = 5
for i in range(dimension, 0, -1):
    print(' ' * (dimension - i) + '*' * (2 * i - 1))
```

Exercice 6: Toujours avec une **boucle for**, dessiner le motif suivant en utilisant des **espaces ' '** et des **étoiles '*'**:

```python
'''
    *
   ***
  *****
 *******
*********
 *******
  *****
   ***
    *
'''
dimension = 5
for i in range(1, dimension + 1):
    print(' ' * (dimension - i) + '*' * (2 * i - 1))
for i in range(dimension - 1, 0, -1):
    print(' ' * (dimension - i) + '*' * (2 * i - 1))
```
